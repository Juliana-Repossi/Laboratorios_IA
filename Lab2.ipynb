###Questão 4

Apresente a soma dos valores, os valores mínimos e máximos, a média e o desvio padrão de cada
coluna do dataset do exercício 3. Apresente também o intervalo de confiança de cada característica.

```python
print(data_frame_filtro)

print("\nSoma:")
print(data_frame_filtro.sum())

print("\nMin:")
print(data_frame_filtro.min())

print("\nMax:")
print(data_frame_filtro.max())

print("\nMédia:")
print(data_frame_filtro.mean())

print("\nDesvio Padrão:")
print(data_frame_filtro.std())

from scipy import stats
from numpy import sqrt

column_b = data_frame_filtro[["b"]]
print ('\n95% intervalo de confiança caracteristica b:',
       stats.norm.interval(0.95, loc=column_b.mean(), 
                           scale=column_b.std()/sqrt(len(column_b))))

column_e = data_frame_filtro[["e"]]
print ('\n95% intervalo de confiança caracteristica e:',
       stats.norm.interval(0.95, loc=column_e.mean(), 
                           scale=column_e.std()/sqrt(len(column_e))))

column_f = data_frame_filtro[["f"]]
print ('\n95% intervalo de confiança caracteristica f:',
       stats.norm.interval(0.95, loc=column_f.mean(), 
                           scale=column_f.std()/sqrt(len(column_f))))

print ('\n95% intervalo de confiança das caracteristicas b,e,f:',
       stats.norm.interval(0.95, loc=data_frame_filtro.mean(), 
                           scale=data_frame_filtro.std()/sqrt(len(data_frame_filtro))))

# Ou essa função que já descreve quase tudo
# count - numero de observações não nulas
# max , min , mean
# std - desvio padrão
# percentis (25, 50 , 75)
data_frame_filtro.describe()
```

###Questão 10

Apresente um gráfico com a projeção TSNE criado no exercício 1

```python
#data frame de numeros aleatorios
copy_data = data_frame.copy()

#adicionando classes ao data frame ----{valor inicial, final e qtd de num para gerar}
copy_data['class'] = np.random.randint(1,5,copy_data.shape[0])

#print(copy_data)


#separar informação de classe - colunas de  'a' a 'l'
data_train = copy_data.loc [:,'a':'l']

data_train
```

```python
from sklearn.manifold import TSNE

tsne = TSNE()


view = tsne.fit_transform(data_train)

#diminuiu a matriz de de (200,12) - (200,2)
#printpd.DataFrame(view))
#print(data_train.shape)
#print(view.shape)

#o que seria um grafico de dispersão com todas caracteristicas
#sns.scatterplot(copy_data)

sns.scatterplot(x=view[:,0],y=view[:,1],hue=copy_data['class'])
```

As classes não possuem distinção porque foram usados dados aleatorios

Vamos adicionar uma característica que diferencia os dados

```python
conditions = [
    (copy_data['class']==1),
    (copy_data['class']==2),
    (copy_data['class']==3),
    (copy_data['class']==4)
]

difer = [np.random.randint(1,4),
         np.random.randint(2,7),
         np.random.randint(10,13),
         np.random.randint(30,100)
         ]

copy_data_train = data_train.copy()
#adicionar coluna
copy_data_train['m'] = np.select(conditions,difer)
copy_data_train
```

```python
view2 = tsne.fit_transform(copy_data_train)

#print(view2)

sns.scatterplot(x=view2[:,0],y=view2[:,1],hue=copy_data['class'])
```
